```python
class SyntheticData[T](Dataset):
    """A small synthetic segmentation dataset.

    It is a sequence dataset of 5000 tuples (image, class, mask), where:
    - image: before transformation, an RGB PIL Image.
    - class: int 0..4, the label index.
    - mask:  numpy array of dtype=bool, shape (H, W), same size as image.
    """

    CLASSES: Final[tuple[str, ...]] = ("circle", "square", "triangle", "star", "diamond")

    def __init__(
        self,
        path: Path,
        split: Literal["train", "val", "test"],
        transform: Callable[[PIL.Image.Image], T],
    ) -> None:
        self.path = path
        self.transform = transform
        self.split = split

        self.class_to_idx = {class_name: i for i, class_name in enumerate(self.CLASSES)}

        images = sorted((self.path / "images" / self.split).glob("*.png"))
        masks = sorted((self.path / "masks" / self.split).glob("*.png"))
        labels = sorted((self.path / "labels" / self.split).glob("*.txt"))

        assert images, f"No images found in {self.path / 'images' / self.split}"
        assert len(images) == len(masks) == len(labels), (
            "Number of images, masks, and labels must be the same"
        )
        assert [p.stem for p in images] == [p.stem for p in masks] == [p.stem for p in labels], (
            "Image/mask/label filename mismatch."
        )
        self.image_names = [p.stem for p in images]

    def __len__(self) -> int:
        return len(self.image_names)

    def __getitem__(self, idx) -> tuple[T, int, np.ndarray]:
        name = self.image_names[idx]
        img_path = self.path / "images" / self.split / f"{name}.png"
        mask_path = self.path / "masks" / self.split / f"{name}.png"
        label_path = self.path / "labels" / self.split / f"{name}.txt"

        img = PIL.Image.open(img_path).convert("RGB")
        mask = np.array(PIL.Image.open(mask_path), dtype=bool)
        label = self.class_to_idx[label_path.read_text().strip()]

        img_transformed: T = self.transform(img)

        return img_transformed, label, mask


def show_image_row(
    image_dict: dict[str, PIL.Image.Image | np.ndarray | Tensor], size: float = 3.0
) -> None:
    n = len(image_dict)
    _, axs = plt.subplots(1, n, figsize=(size * n, size), constrained_layout=True, squeeze=True)
    if n == 1:
        axs = [axs]
    for ax, (title, img) in zip(axs, image_dict.items(), strict=True):
        ax.imshow(img)
        ax.set_title(title)
        ax.axis("off")
    plt.show()
```

## Training a classifier [do not modify]

```python
def train(
    model: nn.Module,
    train_loader: torch.utils.data.DataLoader,
    val_loader: torch.utils.data.DataLoader,
    test_loader: torch.utils.data.DataLoader,
    epochs: int = 5,
    **optimizer_kwargs: Any,
) -> None:
    print(f"ðŸš€ Training CNN for {epochs} epochs...")
    optimizer = optim.AdamW(model.parameters(), **optimizer_kwargs)

    for epoch in range(epochs):
        _train_epoch(model, train_loader, optimizer, desc=f"Epoch {epoch + 1}/{epochs} training  ")
        val_metrics = evaluate(model, val_loader, desc=f"Epoch {epoch + 1}/{epochs} validation")
        print(
            f"Epoch {epoch + 1}/{epochs} â€” "
            + f"val loss: {val_metrics['loss']:.3f}, val acc: {val_metrics['accuracy']:.1%}"
        )

    test_metrics = evaluate(model, test_loader, desc="Test Evaluation")
    print(
        "âœ… Model training complete: "
        + f"Test loss: {test_metrics['loss']:.3f}, test acc: {test_metrics['accuracy']:.1%}"
    )


def _train_epoch(
    model: nn.Module,
    dataloader: torch.utils.data.DataLoader,
    optimizer: optim.Optimizer,
    desc: str,
) -> dict[str, float]:
    model.train()
    device = next(model.parameters()).device
    total_loss = 0
    total_correct = 0
    total_samples = 0
    progress_bar = tqdm(dataloader, desc=desc)
    for imgs, labels, _ in progress_bar:
        imgs, labels = imgs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(imgs)
        loss = nn.CrossEntropyLoss()(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

        _, predicted = torch.max(outputs.data, dim=1)
        total_samples += labels.shape[0]
        total_correct += (predicted == labels).sum().item()

        progress_bar.set_postfix(
            train_loss=f"{total_loss / (total_samples / labels.shape[0]):.3f}",
            train_acc=f"{total_correct / total_samples:.1%}",
        )

    return {"loss": total_loss / len(dataloader), "accuracy": total_correct / total_samples}


def evaluate(
    model: nn.Module, dataloader: torch.utils.data.DataLoader, desc: str
) -> dict[str, float]:
    model.eval()
    device = next(model.parameters()).device
    total_loss = 0
    total_correct = 0
    total_samples = 0
    with torch.no_grad():
        for imgs, labels, _mask in tqdm(dataloader, desc=desc):
            imgs, labels = imgs.to(device), labels.to(device)
            outputs = model(imgs)
            loss = nn.CrossEntropyLoss()(outputs, labels)
            total_loss += loss.item()
            _, predicted = torch.max(outputs.data, dim=1)
            total_samples += labels.shape[0]
            total_correct += (predicted == labels).sum().item()

    return {"loss": total_loss / len(dataloader), "accuracy": total_correct / total_samples}


class DataloaderArgs(TypedDict, total=False):
    batch_size: int
    shuffle: bool
    num_workers: int
    pin_memory: bool
```

## GradCAM implementation (add your code)

```python
class GradCAM:
    """
    Pure PyTorch implementation of Grad-CAM.

    Usage:
    grad_cam = GradCAM(model=model, target_layers=[layer1, layer2])
    grayscale_cam = grad_cam(input_tensor, targets=[class_id])
    """

    def __init__(self, model: nn.Module, target_layers: Iterable[nn.Module]) -> None:
        """
        Args:
            model: The PyTorch model to interpret.
            target_layers: A list/iterable of layers (nn.Module) where the CAM 
                           should be calculated (e.g., the last convolutional layer).
        """
        self.model = model
        self.model.eval()
        self.target_layers = list(target_layers)
        
        # Storage for activations and gradients
        self.activations = {}
        self.gradients = {}
        self.handles = []

        # Register hooks
        for layer in self.target_layers:
            self.handles.append(layer.register_forward_hook(self._get_activation_hook(layer)))
            self.handles.append(layer.register_full_backward_hook(self._get_gradient_hook(layer)))

    def _get_activation_hook(self, layer):
        def hook(module, input, output):
            self.activations[layer] = output.detach()
        return hook

    def _get_gradient_hook(self, layer):
        def hook(module, grad_input, grad_output):
            self.gradients[layer] = grad_output[0].detach()
        return hook

    def _normalize(self, heatmap: torch.Tensor) -> torch.Tensor:
        """
        Normalizes a heatmap (B, 1, H, W) to 0..1 per image in the batch.
        """
        B, C, H, W = heatmap.shape
        flat_map = heatmap.view(B, -1)
        
        mins = flat_map.min(dim=1, keepdim=True)[0].view(B, 1, 1, 1)
        maxs = flat_map.max(dim=1, keepdim=True)[0].view(B, 1, 1, 1)
        
        denominator = maxs - mins
        denominator[denominator == 0] = 1.0
        
        normalized = (heatmap - mins) / denominator
        return normalized

    def __call__(self, input_tensor: torch.Tensor, targets: Iterable[int] | None = None) -> np.ndarray:
        """
        Args:
            input_tensor: Input batch (B, C, H, W) normalized for the model.
            targets: List of target class indices for each image in the batch. 
                     If None, uses the top-1 prediction.
        
        Returns:
            numpy array of shape (B, H, W) containing the Grad-CAM heatmaps, values 0..1.
        """
        input_tensor.requires_grad_(True)
        
        self.model.zero_grad()
        output = self.model(input_tensor)
        
        if targets is None:
            # If no targets provided, grab the index of the highest score
            targets = output.argmax(dim=1).tolist()
        
        if len(targets) != len(input_tensor):
            raise ValueError("Number of targets must match batch size.")

        target_scores = torch.gather(output, 1, torch.tensor(targets, device=input_tensor.device).view(-1, 1))
        loss = torch.sum(target_scores)
        
        loss.backward()

        aggregated_heatmap = torch.zeros(
            (input_tensor.size(0), 1, input_tensor.size(2), input_tensor.size(3)),
            device=input_tensor.device, 
            dtype=torch.float32
        )

        for layer in self.target_layers:
            grads = self.gradients[layer] 
            acts = self.activations[layer]
            
            weights = torch.mean(grads, dim=(2, 3), keepdim=True)
            
            weighted_acts = torch.sum(weights * acts, dim=1, keepdim=True)
            
            cam = F.relu(weighted_acts)
            
            cam_resized = F.interpolate(
                cam, 
                size=(input_tensor.size(2), input_tensor.size(3)), 
                mode='bilinear', 
                align_corners=False
            )
            
            cam_normalized = self._normalize(cam_resized)
            
            aggregated_heatmap += cam_normalized

        avg_heatmap = aggregated_heatmap / len(self.target_layers)

        self.activations.clear()
        self.gradients.clear()

        return avg_heatmap.squeeze(1).cpu().detach().numpy()
    
    def __del__(self):
        for handle in self.handles:
            handle.remove()
```

```python
def heatmap_to_rgb_image(
    heatmap: np.ndarray, min: float | None = None, max: float | None = None
) -> PIL.Image.Image:
    """
    Converts a single-channel heatmap to an RGB pillow image using a colormap.

    Args:
    - heatmap: shape (H, W), will be normalized by mapping min..max to 0..1.
    - min: minimum value for normalization, defaults to heatmap.min().
    - max: maximum value for normalization, defaults to heatmap.max()
    """
    heatmap = heatmap.astype(np.float32)
    if min is None:
        min = heatmap.min()
    if max is None:
        max = heatmap.max()
    heatmap = (heatmap - min) / (max - min + 1e-8)
    heatmap_uint8 = (np.clip(heatmap, 0.0, 1.0) * 255).astype(np.uint8)
    heatmap_color = cv2.applyColorMap(heatmap_uint8, cv2.COLORMAP_JET)
    heatmap_color = cv2.cvtColor(heatmap_color, cv2.COLOR_BGR2RGB)
    return PIL.Image.fromarray(heatmap_color)
```

```python
class BasicSamPipeline:
    def __call__(self, images: Tensor) -> Tensor:
        """
        Input: normalized images, shape (B, C=3, H, W).
        Output: masks tensor of shape (B, H, W), dtype=bool.
        """
        B, C, H, W = images.shape
        # The basic pipeline always uses a single center point for each image.
        point_coords = np.array([[(W // 2, H // 2)] for _ in range(B)])

        # The basic pipeline always uses a single foreground point, no background points.
        point_labels = np.array([[1] for _ in range(B)], dtype=np.int64)

        return self.segment(images, point_coords, point_labels)

    def segment(
        self, images: Tensor, point_coords: np.ndarray, point_labels: np.ndarray
    ) -> Tensor:
        """
        Args:
        - images: normalized images, shape (B, C=3, H, W).
        - point_coords: point coordinates within each image, shape (B, num_points, 2), format (x,y).
            Note the format is not (h,w)=(y,x), but (x,y)!
        - point_labels: point labels, shape (B, num_points), dtype int64.
            Label 1 is foreground (should be in mask), 0 is background (shouldn't be in mask).

        Returns: segmentation masks, shape (B, H, W), dtype=bool.
        """
        B, C, H, W = images.shape
        assert C == 3, f"Expected images.shape=(B, C=3, H, W), got: {images.shape}"
        num_points = point_coords.shape[1]
        assert point_coords.shape == (B, num_points, 2), f"Expected point_coords.shape=({B=}, num_points, 2), got: {point_coords.shape}"
        assert point_labels.shape == (B, num_points), f"Expected point_labels.shape=({B=}, num_points), got: {point_labels.shape}"

        results = list[Tensor]()
        for image, pt_coords, pt_labels in zip(images, point_coords, point_labels, strict=True):
            sam_predictor.set_image(np.array(inverse_transform(image)))
            masks, scores, _logits = sam_predictor.predict(
                point_coords=pt_coords, point_labels=pt_labels, multimask_output=True
            )
            best_mask = masks[np.argmax(scores)]
            results.append(torch.tensor(best_mask, dtype=torch.bool))
        return torch.stack(results)
```

```python
class SamForegroundPipeline(BasicSamPipeline):
    def __call__(self, images: Tensor) -> Tensor:
        """
        Input: normalized images, shape (B, C=3, H, W).
        Output:
        1) masks tensor of shape (B, H, W), dtype=bool.
        """
        B, C, H, W = images.shape
        offset = 1

        grad_cam = GradCAM(model=model, target_layers=target_layers)
        cams = grad_cam(images.to(device), targets=None)
        point_coords = []
        for cam in cams:
            cam_resized = cv2.resize(cam, (W, H))
            # Find the highest activation point
            y, x = np.unravel_index(np.argmax(cam_resized), cam_resized.shape)
            y, x = scipy.ndimage.center_of_mass(cam_resized > 0.8 * cam_resized.max())
            y, x = int(np.round(y)), int(np.round(x))
            points = [(x, y), (x + offset, y + offset), (x - offset, y - offset),
                      (x + offset, y - offset), (x - offset, y + offset), 
                      (x + offset, y), (x - offset, y),
                      (x, y + offset), (x, y - offset)]
            # Clamp coordinates to image bounds
            valid_points = []
            for px, py in points:
                px = min(max(px, 0), W - 1)
                py = min(max(py, 0), H - 1)
                valid_points.append((px, py))
            point_coords.append(valid_points)


        point_coords = np.array(point_coords)

        point_labels = np.array([[1]*9 for _ in range(B)], dtype=np.int64)
        
        results = self.segment(images, point_coords, point_labels)
        
        # To preserve compatibility with previous version
        self.point_coords = point_coords
        self.point_labels = point_labels
        
        return results
```

```python
class SamBackgroundPipeline(BasicSamPipeline):
    def __call__(self, images: Tensor) -> Tensor:
        """
        Input: normalized images, shape (B, C=3, H, W).
        Output: 
          - masks: (B, H, W) bool
        """
        B, C, H, W = images.shape
        offset = 1
        bg_offset = 5

        grad_cam = GradCAM(model=model, target_layers=target_layers)
        cams = grad_cam(images.to(device), targets=None)
        
        final_coords = []
        final_labels = []
        
        for i in range(B):
            cam = cv2.resize(cams[i], (W, H))
            
            cam_resized = cv2.resize(cam, (W, H))
            y, x = scipy.ndimage.center_of_mass(cam_resized > 0.8 * cam_resized.max())
            y, x = int(np.round(y)), int(np.round(x))
            fg_points = [(x, y), (x + offset, y + offset), (x - offset, y - offset),
                      (x + offset, y - offset), (x - offset, y + offset), 
                      (x + offset, y), (x - offset, y),
                      (x, y + offset), (x, y - offset)]
            # Clamp coordinates to image bounds
            valid_fg_points = []
            for px, py in fg_points:
                px = min(max(px, 0), W - 1)
                py = min(max(py, 0), H - 1)
                valid_fg_points.append((px, py))
            fg_points = valid_fg_points
            
            cutoff_point = cam_resized.min() + 0.2 * (cam_resized.max() - cam_resized.min())

            # Find background point using center of mass of low-activation areas
            by, bx = scipy.ndimage.center_of_mass(cam_resized < cutoff_point)
            by, bx = int(np.round(by)), int(np.round(bx)) 

            if cam_resized[by, bx] > cutoff_point:
                # If the center of mass is not below the cutoff, pick a point further away
                candidates = np.argwhere(cam_resized <= cutoff_point)
                dists = np.linalg.norm(candidates - np.array([y, x]), axis=1)
                farthest_idx = np.argmax(dists)
                by, bx = candidates[farthest_idx]
                by, bx = int(by), int(bx)
            bg_points = [(bx, by)]

            points = fg_points + bg_points
            
            # Labels: 1 for foreground, 0 for background
            labels = [1] * len(fg_points) + [0] * len(bg_points)
            
            final_coords.append(points)
            final_labels.append(labels)

        
        final_coords = np.array(final_coords)
        final_labels = np.array(final_labels, dtype=np.int64)

        # Run SAM with the generated points
        results = self.segment(images, final_coords, final_labels)

        # To preserve compatibility with previous version
        self.point_coords = final_coords
        self.point_labels = final_labels
        
        return results
```

```python
def sam_example_custom_pipeline(use_pipeline_1 : bool = True, num_imgs : int = None, plot : bool = False):
    images, labels, gt_masks = next(iter(test_loader))
    
    if num_imgs is not None:
        indices = [i for i in range(num_imgs)]
        images, labels, gt_masks = images[indices], labels[indices], gt_masks[indices]
    
    if use_pipeline_1:
        pipeline = SamForegroundPipeline()
    else:
        pipeline = SamBackgroundPipeline()

    results = pipeline(images)
    point_coords = pipeline.point_coords
    point_labels = pipeline.point_labels
    results = results.cpu()

    avg_fg_hit_rate = 0.0
    avg_bg_hit_rate = 0.0
    avg_distance = 0.0
    avg_IoU = 0.0

    num_imgs = len(images)


    # Calculation of metrics
    for i in range(len(images)):
        gt_mask = gt_masks[i].numpy()
        pred_mask = results[i].numpy()

        # We only consider foreground points for hit rate and distance
        fg_points = point_coords[i][point_labels[i] == 1]
        bg_points = point_coords[i][point_labels[i] == 0]
        total_hits_fg = 0 
        total_hits_bg = 0
        total_distance = 0.0

        foreground_center_of_mass = scipy.ndimage.center_of_mass(gt_mask)
        if np.isnan(foreground_center_of_mass[0]) or np.isnan(foreground_center_of_mass[1]):
            # No foreground in GT mask
            num_imgs -= 1
            continue

        fg_y, fg_x = foreground_center_of_mass

        for (x, y) in fg_points:
            if gt_mask[y, x]:
                total_hits_fg += 1
            dist = np.sqrt((x - fg_x) ** 2 + (y - fg_y) ** 2)
            total_distance += dist
        
        for (x, y) in bg_points:
            if not gt_mask[y, x]:
                total_hits_bg += 1
        hit_rate = total_hits_fg / len(fg_points)
        avg_fg_hit_rate += hit_rate
        avg_distance += total_distance / len(fg_points)

        if len(bg_points) > 0:
            hit_rate_bg = total_hits_bg / len(bg_points)
            avg_bg_hit_rate += hit_rate_bg

        intersection = np.logical_and(gt_mask, pred_mask).sum()
        union = np.logical_or(gt_mask, pred_mask).sum()
        IoU = intersection / union if union != 0 else 1.0
        avg_IoU += IoU
    
    if num_imgs == 0:
        print("No valid images for evaluation.")
        return
    
    avg_fg_hit_rate /= num_imgs
    avg_bg_hit_rate /= num_imgs
    avg_distance /= num_imgs
    avg_IoU /= num_imgs

    print(f"Average Hit Rate (Foreground): {avg_fg_hit_rate:.3f}")
    if not use_pipeline_1:
        print(f"Average Hit Rate (Background): {avg_bg_hit_rate:.3f}")
    print(f"Average Distance to GT Mask: {avg_distance:.3f} pixels")
    print(f"Average IoU: {avg_IoU:.3f}")


    if plot:
        for image, result, gt_mask in zip(images, results, gt_masks, strict=True):
            show_image_row(
                {
                    "Input image": inverse_transform(image),
                    "Basic SAM predicted mask": result,
                    "Ground-truth mask": gt_mask,
                }
            )
```
